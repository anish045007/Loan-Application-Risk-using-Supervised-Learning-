### **3.2.1.1 Supervised Machine Learning Classification Algorithm - Decision Tree (Base Model)**

In this section, we focus on implementing the decision tree algorithm as the base model for our risk assessment analysis. The decision tree algorithm is chosen due to its interpretability and ability to handle both categorical and numerical data effectively.

**Metrics Used:**

1. **Gini Coefficient:** The Gini coefficient measures the impurity of a node in the decision tree. A lower Gini coefficient indicates a more homogeneous node, which is desirable for effective classification.

2. **Entropy:** Entropy quantifies the randomness or uncertainty in a node's class distribution. Minimizing entropy helps in creating more pure and informative decision tree nodes.

Our objective is to optimize the decision tree model's performance by fine-tuning hyperparameters and pruning to prevent overfitting. Through this analysis, we aim to achieve a robust and accurate classification of loan applicants into risk segments.


### **3.2.1.2. Supervised Machine Learning Classification Algorithms - Logistic Regression, Support Vector Machine, K Nearest Neighbour (Comparison Models)**

In this section, we compare the performance of three supervised machine learning classification algorithms: Logistic Regression, Support Vector Machine (SVM), and K Nearest Neighbour (KNN). These models are evaluated based on their ability to accurately classify loan applicants into risk segments.

**Metrics Used:**
1. **Accuracy:** Measures the overall correctness of predictions made by the model.
2. **Recall:** Evaluates the model's ability to correctly identify positive cases among all actual positive cases.
3. **Precision:** Assesses the model's accuracy in predicting positive cases among all predicted positive cases.
4. **F1-Score:** Harmonic mean of precision and recall, providing a balanced assessment of model performance.

**Implementation Details:**

# **5. Managerial Insights:**

## **5.1. Appropriate Model: Compare and Contrast {Decision Tree | Logistic Regression | Support Vector Machine | K Nearest Neighbour}**
   
   - **Decision Tree:**
     - Pros: Decision trees offer a straightforward and interpretable representation of the data's decision boundaries. They can handle both numerical and categorical data and are robust to outliers.
     - Cons: They tend to overfit when the tree depth is not properly tuned, leading to suboptimal generalization on unseen data.

   - **Logistic Regression:**
     - Pros: Logistic regression is a simple yet powerful algorithm for binary classification. It provides interpretable coefficients that indicate the impact of each feature on the outcome.
     - Cons: It assumes a linear relationship between features and the target variable, which may not hold true for complex data patterns.

   - **Support Vector Machine (SVM):**
     - Pros: SVMs are effective in high-dimensional spaces and can capture complex relationships through the use of different kernel functions.
     - Cons: They are sensitive to the choice of kernel and regularization parameters, which need careful tuning for optimal performance.

   - **K Nearest Neighbour (KNN):**
     - Pros: KNN is a non-parametric, instance-based learning algorithm that makes predictions based on the similarity of data points. It can capture non-linear relationships and adapt well to varying datasets.
     - Cons: KNN's performance can be sensitive to the choice of the number of neighbors (k) and the distance metric used, requiring experimentation for optimal results.

## **5.2. Relevant or Important Variables or Features (Given the Appropriate Model - K Nearest Neighbour)**
